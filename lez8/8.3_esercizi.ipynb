### Es 1
Hai a disposizione un file `data.csv` contenente dati mensili di traffico aereo con due colonne:

- `date`: data in formato `YYYY-MM` (mese/anno)
- `passengers`: numero di passeggeri per quel mese


Costruisci un modello di **regressione polinomiale** che approssima l’andamento del numero di passeggeri nel tempo.

1. Carica il dataset.
2. Convertilo in un formato numerico utilizzando una colonna `mese_numerico` che conti i mesi a partire da gennaio 1949.
3. Applica una regressione polinomiale (grado a tua scelta).
4. Calcola l’RMSE tra i valori reali e quelli predetti.
5. Visualizza i dati reali e la curva stimata con Plotly.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
import plotly.graph_objects as go
from sklearn.metrics import root_mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("../data.csv")
df.head()
```

```python
df = df.dropna()
```

```python
df['mese_numerico'] = range(1, len(df) + 1)
```

```python
df.head()
```

```python
X = df[["mese_numerico"]]
y = df[["passengers"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, shuffle=False)
```

```python
degree = 3
model = make_pipeline(
    PolynomialFeatures(degree=degree),
    LinearRegression()
)

model.fit(X_train, y_train)
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

rmse_train = root_mean_squared_error(y_train, y_pred_train)
rmse_test = root_mean_squared_error(y_test, y_pred_test)

print("Train RMSE:", rmse_train)
print("Test RMSE:", rmse_test)
```

```python
max_degree = 5
for i in range(1, max_degree+1):
    print(f"\nDegree: {i}")
    model_test_degr = make_pipeline(
        PolynomialFeatures(degree=i),
        LinearRegression()
    )

    model_test_degr.fit(X_train, y_train)
    y_pred_train = model_test_degr.predict(X_train)
    y_pred_test = model_test_degr.predict(X_test)

    rmse_train = root_mean_squared_error(y_train, y_pred_train)
    rmse_test = root_mean_squared_error(y_test, y_pred_test)

    print("Train RMSE:", rmse_train)
    print("Test RMSE:", rmse_test)
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=X_train["mese_numerico"],
    y=y_train["passengers"],
    mode="markers",
    name="Train"
))

fig.show()
```

```python
fig.add_trace(go.Scatter(
    x=X_test["mese_numerico"],
    y=y_test["passengers"],
    mode="markers",
    name="Test"
))

fig.show()
```

```python
fig.add_trace(go.Scatter(
    x=df["mese_numerico"],
    y=model.predict(X).ravel(),
    mode="lines",
    name=f"Regressione polinomiale grado {degree}"
))


fig.show()
```
